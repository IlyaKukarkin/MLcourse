```python
data2000['angel_dummy'] = data2000['angel'] !=0
data2000['seed_dummy'] = data2000['seed'] !=0
data2000['grant_dummy'] = data2000['grant'] !=0
```

```python
df = data2000[['funding_rounds',' funding_total_usd ', 'status', 'angel_dummy','seed_dummy', 'grant_dummy', ' market ', 'country_code' ]].copy()
```

```python
from sklearn import preprocessing
encoder = preprocessing.LabelEncoder()
df = df.dropna()
df[' market '] = encoder.fit_transform(df[' market '])
df['country_code'] = encoder.fit_transform(df['country_code'])

df[' funding_total_usd '] = df[' funding_total_usd '].str.replace(',', '')
df[' funding_total_usd '] = df[' funding_total_usd '].str.replace('-', '')
df[' funding_total_usd '] = df[' funding_total_usd '].str.replace(' ', '')
df[' funding_total_usd '] = pd.to_numeric(df[' funding_total_usd '], errors='coerce')
df = df.dropna()
y = df['status'] != 'operating'
df = df.drop(columns=['status'])
print(df.head())
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df, y, test_size=0.2)
clf = DecisionTreeClassifier(random_state=0, max_depth=7)
clf.fit(X_train, y_train)
clf.predict(X_test)
score = clf.score(X_test, y_test)
print(score)
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(random_state=0, max_depth=13)
clf.fit(X_train, y_train)
clf.predict(X_test)
score = clf.score(X_test, y_test)
print(score)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
clf = GradientBoostingClassifier(random_state=0, learning_rate=0.01, n_estimators=250)
clf.fit(X_train, y_train)
clf.predict(X_test)
score = clf.score(X_test, y_test)
print(score)
```

```python
import xgboost as xgb
clf=xgb.XGBClassifier(random_state=0, max_depth=4, learning_rate = 0.1)
clf.fit(X_train, y_train)
clf.predict(X_test)
score = clf.score(X_test, y_test)
print(score)
```
